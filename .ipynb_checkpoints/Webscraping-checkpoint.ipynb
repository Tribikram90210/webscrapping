```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.worldometers.info/gdp/australia-gdp/#:~:text=Nominal%20(current)%20Gross%20Domestic%20Product,when%20Real%20GDP%20was%20%241%2C524%2C320%2C000%2C000.'
```

```python
html = requests.get(url).content
```

```python
print(html)
```

```python
soup = BeautifulSoup(html,'html.parser')
print(soup) # html.parser sorts the html code into a tree like structure
```

```python
table_body = soup.find('tbody')
```

```python
table_rows = table_body.find_all('tr')
```

```python
table_rows[0]
```

```python
dummy = table_rows[0]
dummy.find_all('td')
```

```python
dummy = table_rows[0]
[items.text for items in dummy.find_all('td')]
```

```python
dummy = table_rows[5]
[items.text for items in dummy.find_all('td')]
```

```python
data = [[items.text for items in table_row.find_all('td')] for table_row in table_rows ]
```

```python
print(data)
```

```python
for table_row in table_rows:
    for item in table_row.find_all('td'):
        print(item.text)
    break

```

```python
import pandas as pd
df = pd.DataFrame(data)
df
```
